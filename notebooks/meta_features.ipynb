# Visualization of meta-features

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from pathlib import Path
```

```python
features = pd.read_parquet(str(Path(os.getcwd()).parent) + '/src/artifacts/meta_features.parquet')
```

```python
features
```

```python
features.keys()
```

```python
drop = ['g_mean.mean', 'g_mean.sd', 'h_mean.mean', 'h_mean.sd', 'nr_cat', 'cat_to_num', 'num_to_cat', 'sd_ratio']
features = features.drop(columns=drop)
```

```python
df = features.copy()
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].replace([np.inf, -np.inf], np.nan)
df[num_cols] = df[num_cols].fillna(df[num_cols].median()) # impute with medians for simple visualization
df.hist(bins=10, figsize=(20, 12))
plt.tight_layout()
plt.show()
```
